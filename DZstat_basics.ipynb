# Задание 2. Работа с выбросами

```python
# Возраст лошади
q1 = result['Возраст лошади'].quantile(0.25)
q3 = result['Возраст лошади'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
age_outliers = result[result['Возраст лошади'].between(lower_bound, upper_bound, inclusive=True)].sort_values('Возраст лошади')
age_outliers
age_outliers_ = pd.concat([result,age_outliers]).drop_duplicates(keep = False)
"""Выбросы в данном случае образовались я думаю в результате человеческого фактора(ошибка ввода данных), 
так как 9 стоит из за опечатки(возможноэто реальный возраст лошади) , по легенде может быть только два значения
1 - Взрослая лошадь и 2- молодая лошадь,
я считаю надо 9 поменять на значение 1 - замену выполнила при работе с пропусками см. ниже"""
age_outliers_.head()
```

```python
# Ректальная температура
"""1. Вышли позиции в которых отсутствуют значения ректальной температуры, которые в альнейшем заменим на медианные значения
   2. вышла позиция с температурой 39,9, и такая температура у Лошади может быть"""
q1 = result['ректальная температура'].quantile(0.25)
q3 = result['ректальная температура'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
rectal_temperature_outliers = result[result['ректальная температура'].between(lower_bound, upper_bound, inclusive=True)].sort_values('ректальная температура')
rectal_temperature_outliers
rectal_temperature_outliers_ = pd.concat([result, rectal_temperature_outliers]).drop_duplicates(keep = False)
rectal_temperature_outliers_.head()
```

```python
# пульс
"""1. Выбросы пульса выявлены у лошадей с значением возраста - 9 (тоже выброс для показателя возраст лошади)
   2. вышли позиции где отсутствуют значения пульса и большая часть остальных показаний также отсутствует
   в 1-м случае высокий пульс прослеживается у лошадей с высокой ректальной температурой, кроме того 
   сталкиваясь с ситуацией "дерись или убегай", симпатическая нервная система выделяет эпинефрин (адреналин),
   который пересиливает тонус блуждающего нерва и может быстро поднять частоту пульса лошади до 220-240 уд/мин. 
   Соотвественно такой пульс имеет место быть.
   Во 2-м случае, я думаю можно вместо отсутствующих значений поставить медианные значения"""
q1 = result['пульс'].quantile(0.25)
q3 = result['пульс'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
pulse_outliers = result[result['пульс'].between(lower_bound, upper_bound, inclusive=True)].sort_values('пульс')
pulse_outliers
pulse_outliers_ = pd.concat([result, pulse_outliers]).drop_duplicates(keep = False)
pulse_outliers_.head()
```

```python
for col in result.columns:
    pct_missing = result[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```
